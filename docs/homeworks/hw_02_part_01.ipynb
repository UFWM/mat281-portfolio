```python
# Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from os import listdir
from os.path import isfile, join

# Importar herramientas de Scikit-learn
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Configuración de gráficos
%matplotlib inline
sns.set_palette("deep", desat=0.6)
sns.set(rc={'figure.figsize': (11.7, 8.27)})

# Cargar y preparar los datos
df_districts = pd.read_csv("districts_info.csv")
df_products = pd.read_csv("products_info.csv").set_index('LP ID')
```

Ahora, queremos sacar la informacion del desglose de conexiones a sitios web durante el 2020, para cada distrito en nuestra dataframe. Para esto, transformemos los distintos IDs de distrito en el formato del nombre de los archivos en los que se encuentran.

```python
districts=df_districts_filter['district_id'].unique().tolist()
districts_path=[str(x)+".csv" for x in districts]
districts_path
```

Luego, podemos para cada uno de estos archivos generar su respectiva dataframe, para luego concatenarlas (manteniendo registro de donde vienen los datos) y asi solo trabajar con una dataframe. Aseguremonos de considerar solamente los datos que tienen "lp_id" no vacio, pues no nos sirven para poder identificar de que sitio web provienen.

```python
districts_list=[]
i=0

for district_path in districts_path:
    archive_path = os.path.join('engagement_data', district_path)
    df_temp=pd.read_csv(archive_path)
    df_temp['district_id']=districts[i]
    districts_list.append(df_temp)
    i=i+1

df=pd.concat(districts_list).reset_index(drop=True)
df
```

```python
df.info()
```

Veamos la tasa de conección a sitios web durante el año.

```python
df['time']=pd.to_datetime(df['time'])
df['time'].value_counts().sort_index(ascending=True).plot(kind='line',title='Overall connections over time')
```

Veamos las aplicaciones mas populares.
